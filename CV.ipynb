```python
#camera face dedtection
cap=cv2.VideoCapture(0)
face_cascade=cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
eye_cascade=cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_eye.xml")
smile_cascade=cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_smile.xml")
while True:
    ret, frame=cap.read()
    face=face_cascade.detectMultiScale(frame,scaleFactor=1.1,minNeighbors=5)
    for (x,y,w,h) in face:
        label='face'
        label3='smile'
        cv2.putText(frame, label, (x, y-10), 
            cv2.FONT_HERSHEY_SIMPLEX, 
            0.7, (20, 11, 230), 2)
        cv2.rectangle(frame,(x,y),(x+w,y+h),(20,11,230),4)
        sub_face=frame[y:y+h, x:x+w]
        eye=eye_cascade.detectMultiScale(sub_face,scaleFactor=1.3,minNeighbors=10)
        smile=smile_cascade.detectMultiScale(sub_face,scaleFactor=1.7,minNeighbors=10)
        for (ex,ey,ew,eh) in eye:
            cv2.rectangle(sub_face,(ex,ey),(ex+ew,ey+eh),(255,0,0),2)
            label2='eye'
            cv2.putText(sub_face, label2, (ex, ey-10),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.5, (255, 0, 0), 1)
        for (sx,sy,sw,sh) in smile:
            label2='smile'
            cv2.putText(sub_face, label2, (sx, sy-10),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.5, (0, 255, 0), 1)
            cv2.rectangle(sub_face,(sx,sy),(sx+sw,sy+sh),(0,255,0),2)

    cv2.imshow("webcam",frame)

    if cv2.waitKey(1)& 0xFF==ord('a'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
from supervision import Detections
from huggingface_hub import hf_hub_download
from PIL import Image
```

```python
cap=cv2.VideoCapture(0)
model_path = hf_hub_download(repo_id="AdamCodd/YOLOv11n-face-detection", filename="model.pt")
model = YOLO(model_path)

while True:
    ret, img=cap.read()
    results = model(img)
    detections = Detections.from_ultralytics(results[0])
   
   
    for xyxy in detections.xyxy:
        x1,y1,x2,y2=map(int,xyxy)
        cv2.rectangle(img,(x1,y1),(x2,y2),(0,0,255),2)
        cv2.imshow("image",img)
    if cv2.waitKey(1)& 0xFF==ord('a'):
        break

cap.release()
cv2.destroyAllWindows()
```
